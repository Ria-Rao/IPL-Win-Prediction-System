```python
deliveries_df=pd.DataFrame(deliveries)
```

```python
deliveries_df["tot_runs"]=deliveries_df["runs_off_bat"]+deliveries_df["extras"]
```

```python
deliveries_df.head()
```

```python
match_df=pd.DataFrame(match)
```

```python
total_score_df=deliveries_df.groupby(["match_id","innings"]).sum()["tot_runs"].reset_index()
```

```python
total_score_df.tail()
```

```python
total_score_df=total_score_df[total_score_df["innings"]==1]
```

```python
total_score_df
```

```python
match_df=match_df.merge(total_score_df[["match_id","tot_runs"]],left_on="id",right_on="match_id")
```

```python
match_df["team1"].unique()
```

```python
teams=['Gujarat Titans', 'Mumbai Indians', 'Chennai Super Kings',
       'Sunrisers Hyderabad', 'Royal Challengers Bangalore',
       'Lucknow Super Giants', 'Delhi Capitals',
       'Kolkata Knight Riders', 'Rajasthan Royals', 'Kings XI Punjab',]
```

```python
match_df["team1"]=match_df["team1"].str.replace("Delhi Daredevils","Delhi Capitals")
match_df["team2"]=match_df["team2"].str.replace("Delhi Daredevils","Delhi Capitals")

match_df["team2"]=match_df["team2"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
match_df["team1"]=match_df["team1"].str.replace("Deccan Chargers","Sunrisers Hyderabad")

match_df["team1"]=match_df["team1"].str.replace("Punjab Kings","Kings XI Punjab")
match_df["team2"]=match_df["team2"].str.replace("Punjab Kings","Kings XI Punjab")
```

```python
match_df=match_df[match_df["team1"].isin(teams)]
match_df=match_df[match_df["team2"].isin(teams)]
```

```python
match_df
```

```python
match_df=match_df[match_df["dl_applied"]==0]
```

```python
match_df
```

```python
match_df=match_df[["match_id","city","winner","tot_runs"]]
```

```python
match_df
```

```python
delivery_df=match_df.merge(deliveries_df,on="match_id")
```

```python
delivery_df=delivery_df[delivery_df["innings"]==2]
```

```python
delivery_df.shape
```

```python
delivery_df
```

```python
delivery_df["current_score"]=delivery_df.groupby(["match_id"])["tot_runs_y"].cumsum()
```

```python
delivery_df
```

```python
delivery_df["runs_left"]=delivery_df["tot_runs_x"]-delivery_df["current_score"]+1
```

```python
delivery_df
```

```python
delivery_df["player_dismissed"]=delivery_df["player_dismissed"].fillna("0")
delivery_df["player_dismissed"]=delivery_df["player_dismissed"].apply(lambda x:x if x=="0" else "1")
delivery_df["player_dismissed"]=delivery_df["player_dismissed"].astype(int)
wickets=delivery_df.groupby("match_id")["player_dismissed"].cumsum().values
delivery_df["wickets"]=10-wickets
delivery_df.head()
```
